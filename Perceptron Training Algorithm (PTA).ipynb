```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
# step function check if input is positive than return 1 else 0
def stepFunction(t):
    if t >= 0:
        return 1
    return 0

# this function return prediction by X*W + b and apply step function
def prediction(X, W, b):
    return stepFunction((np.matmul(X,W)+b)[0])

# The function should receive as inputs the data X, the labels y,
# the weights W (as an array), and the bias b,
# update the weights and bias W, b, according to the perceptron algorithm,
# and return W and b.
def perceptronStep(X, y, W, b):
    for i in range(len(X)):
        delta_W=y[i]-prediction(X[i],W,b)
        if(delta_W!=0):
          W[0] += delta_W*X[i][0]
          W[1] += delta_W*X[i][1]
          b += delta_W
          break;
    return W, b
    
# This function runs the perceptron algorithm repeatedly on the dataset,
# and returns a few of the boundary lines obtained in the iterations,for plotting purposes.

def trainPerceptronAlgorithm(X, y,num_epochs = 25):
    W = np.array([[random.uniform(0,0.2)],[random.uniform(0,0.2)]]) # generate random values and assign to weight
    b = random.uniform(0,0.2)
    boundary_lines = [] #list to store coeffienct of line
    for i in range(num_epochs):
        W, b = perceptronStep(X, y, W, b)
        boundary_lines.append((-W[0]/W[1], -b/W[1])) # coeffienct of line
    return boundary_lines

# this function works same as perceptronstep but it only has one weight to update for not gate
def perceptronStepForNot(X, y, W, b):
    for i in range(len(X)):
        delta_W=y[i]-prediction(X[i],W,b)
        if(delta_W!=0):
          W[0] += delta_W*X[i]
          b += delta_W
    return W, b

# this function works same as trainperceptronAlgorithm but it only has one weight for not gate
def trainPerceptronAlgorithmForNot(X, y,num_epochs = 25):
    W = np.array([[random.uniform(-3,3)]])
    b = random.uniform(-3,3)
    boundary_lines = []
    for i in range(num_epochs):
        W, b = perceptronStepForNot(X, y, W, b)
        boundary_lines.append((-b/W[0]))
    return boundary_lines
```

### AND Gate

```python
# initialize input and output label X,y
X=np.array([[0, 0],[0, 1],[1,0],[1, 1]])  
y=np.array([0, 0, 0, 1])
z=trainPerceptronAlgorithm(X, y ,num_epochs = 20)
```

```python
#this is count how many epochs function will converge
count=1
for i in range(len(z)-1):
  if(z[i]!=z[i+1]):
    count+=1
  else:
    break;
count
```
